```python
import numpy as np
import pandas as pd
import torch
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('/content/breast-cancer.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id'],inplace=True)
```

```python
df.head()
```

**TRAIN TEST SPLIT**

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

**Scaling or Standardisation**

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

**Label Encoding**

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

**Numpy Array to PyTorch tensors**

```python
X_train_tensor=torch.from_numpy(X_train)
X_test_tensor=torch.from_numpy(X_test)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
y_train_tensor.shape
```

```python
from torch.utils.data import Dataset,DataLoader

class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.features=features
    self.labels=labels
  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self,index):
    return self.features[index],self.labels[index]
```

```python
train_dataset=CustomDataset(X_train_tensor,y_train_tensor)
test_dataset=CustomDataset(X_test_tensor,y_test_tensor)
```

```python
train_dataset[10]
```

```python
train_loader=DataLoader(train_dataset,batch_size=16,shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=16,shuffle=False)
```

**Important parameters**

```python
learning_rate=0.1
epochs=25
```

```python
#built-in loss fn
loss_function=nn.BCELoss()
```

**Training pipeline**

```python
# Ensure all tensors are double precision
X_train_tensor = torch.from_numpy(X_train).double()
X_test_tensor = torch.from_numpy(X_test).double()
y_train_tensor = torch.from_numpy(y_train).double()
y_test_tensor = torch.from_numpy(y_test).double()

#Create model
model = Model(X_train_tensor.shape[1])
model.double() # Cast the model's parameters to double precision


# Built-in optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# Training loop
# Define loop
for epoch in range(epochs):
  for batch_features,batch_labels in train_loader:

    # Forward pass
    y_pred = model(X_train_tensor)

    # Loss calculate
    loss = loss_function(y_pred, y_train_tensor.view(-1, 1))

    # Backward pass
    loss.backward()

    # Parameters update using optimizer
    optimizer.step()
    optimizer.zero_grad()  #manually zero gradients

    print(f'epoch:{epoch+1},Loss:{loss.item()}')
```
